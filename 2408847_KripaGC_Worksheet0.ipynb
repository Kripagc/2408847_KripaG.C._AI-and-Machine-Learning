```python
def convert_units(value, conversion_type):
    """
    Converts a value between different units of measurement.

    Parameters:
        value (float): The number to convert
        conversion_type (str): The type of conversion to perform

    Returns:
        float: The converted value, or None if invalid
    """

    if conversion_type == "1":   # meters to feet
        return value * 3.28084

    elif conversion_type == "2": # feet to meters
        return value / 3.28084

    elif conversion_type == "3": # kg to pounds
        return value * 2.20462

    elif conversion_type == "4": # pounds to kg
        return value / 2.20462

    elif conversion_type == "5": # liters to gallons
        return value * 0.264172

    elif conversion_type == "6": # gallons to liters
        return value / 0.264172

    else:
        return None  # invalid option


def main():
    """Main function that runs the unit converter program."""

    print("===== Unit Converter =====")
    print("Choose a conversion:")
    print("1. Meters → Feet")
    print("2. Feet → Meters")
    print("3. Kilograms → Pounds")
    print("4. Pounds → Kilograms")
    print("5. Liters → Gallons")
    print("6. Gallons → Liters")

    choice = input("Enter your choice (1-6): ")

    try:
        value = float(input("Enter the value to convert: "))

        result = convert_units(value, choice)

        if result is None:
            print("Invalid choice. Please enter a number between 1 and 6.")
        else:
            print(f"Result: {result:.4f}")

    except ValueError:
        print("Invalid input! Please enter a numeric value.")


# Run the program
main()
```

```python
def calculate_sum(numbers):
    """
    Returns the sum of a list of numbers.

    Parameters:
        numbers (list): A list of numeric values

    Returns:
        float: The sum of all numbers in the list
    """
    total = 0
    for n in numbers:
        total += n
    return total


def calculate_average(numbers):
    """
    Returns the average of a list of numbers.

    Parameters:
        numbers (list): A list of numeric values

    Returns:
        float: The average of all numbers in the list
    """
    return calculate_sum(numbers) / len(numbers)


def calculate_maximum(numbers):
    """
    Returns the maximum value in a list of numbers.

    Parameters:
        numbers (list): A list of numeric values

    Returns:
        float: The largest number in the list
    """
    biggest = numbers[0]
    for n in numbers:
        if n > biggest:
            biggest = n
    return biggest


def calculate_minimum(numbers):
    """
    Returns the minimum value in a list of numbers.

    Parameters:
        numbers (list): A list of numeric values

    Returns:
        float: The smallest number in the list
    """
    smallest = numbers[0]
    for n in numbers:
        if n < smallest:
            smallest = n
    return smallest


def main():
    """Main function that runs the math operations program."""

    print("===== Math Operations =====")
    print("Choose an operation:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    choice = input("Enter your choice (1-4): ")

    try:
        raw = input("Enter numbers separated by spaces (e.g. 1 2 3 4): ")
        numbers = [float(x) for x in raw.split()]

        if len(numbers) == 0:
            print("Error: You entered an empty list!")

        elif choice == "1":
            print(f"Sum = {calculate_sum(numbers)}")

        elif choice == "2":
            print(f"Average = {calculate_average(numbers)}")

        elif choice == "3":
            print(f"Maximum = {calculate_maximum(numbers)}")

        elif choice == "4":
            print(f"Minimum = {calculate_minimum(numbers)}")

        else:
            print("Invalid choice! Please enter 1, 2, 3 or 4.")

    except ValueError:
        print("Invalid input! Please enter numeric values only.")


main()
```

```python
# SECTION: List Manipulation (Q.1 to Q.7)
# Q.1 - List Manipulation - Extract Every Other Element
# Takes a list and returns elements at positions 0, 2, 4...
def extract_every_other(lst):
    """
    Extracts every other element starting from the first.

    Parameters:
        lst (list): Input list

    Returns:
        list: Every other element from the original list
    """
    return lst[::2]

print("Q.1:", extract_every_other([1, 2, 3, 4, 5, 6]))
```

```python
# Q.2 - List Manipulation - Slice a Sublist
# Returns elements from start index to end index (inclusive)
def get_sublist(lst, start, end):
    """
    Returns a sublist from start to end index (inclusive).

    Parameters:
        lst (list): Input list
        start (int): Starting index
        end (int): Ending index (inclusive)

    Returns:
        list: Sublist from start to end
    """
    return lst[start:end+1]

print("Q.2:", get_sublist([1, 2, 3, 4, 5, 6], 2, 4))
```

```python
# Q.3 - List Manipulation - Reverse a List Using Slicing
# Uses slicing trick [::-1] to reverse the list
def reverse_list(lst):
    """
    Reverses a list using slicing.

    Parameters:
        lst (list): Input list

    Returns:
        list: Reversed version of the input list
    """
    return lst[::-1]

print("Q.3:", reverse_list([1, 2, 3, 4, 5]))
```

```python
# Q.4 - List Manipulation - Remove First and Last Elements
# Slices from index 1 to second-to-last element
def remove_first_last(lst):
    """
    Removes the first and last elements of a list.

    Parameters:
        lst (list): Input list

    Returns:
        list: List without first and last elements
    """
    return lst[1:-1]

print("Q.4:", remove_first_last([1, 2, 3, 4, 5]))
```

```python
# Q.5 - List Manipulation - Get First N Elements
# Returns only the first n elements using slicing
def get_first_n(lst, n):
    """
    Returns the first n elements of a list.

    Parameters:
        lst (list): Input list
        n (int): Number of elements to return from the start

    Returns:
        list: First n elements
    """
    return lst[:n]

print("Q.5:", get_first_n([1, 2, 3, 4, 5], 3))
```

```python
# Q.6 - List Manipulation - Extract Last N Elements
# Uses negative slicing to grab elements from the end
def get_last_n(lst, n):
    """
    Returns the last n elements of a list.

    Parameters:
        lst (list): Input list
        n (int): Number of elements to return from the end

    Returns:
        list: Last n elements
    """
    return lst[-n:]

print("Q.6:", get_last_n([1, 2, 3, 4, 5], 2))
```

```python
# Q.7 - List Manipulation - Reverse Skip
# Starts from second-to-last element, goes backward, skips one each time
def reverse_skip(lst):
    """
    Returns every second element starting from second-to-last, moving backward.

    Parameters:
        lst (list): Input list

    Returns:
        list: Every second element in reverse order
    """
    return lst[-2::-2]

print("Q.7:", reverse_skip([1, 2, 3, 4, 5, 6]))
```

```python
# SECTION: Nested Lists (Q.1 to Q.8)
# Q.1 - Nested Lists - Flatten a Nested List
# Converts [[1,2],[3,4],[5]] into [1,2,3,4,5]
def flatten(lst):
    """
    Flattens a nested list into a single list.

    Parameters:
        lst (list): A nested list

    Returns:
        list: A single flattened list
    """
    result = []
    for sublist in lst:
        for item in sublist:
            result.append(item)
    return result

print("Q.1:", flatten([[1, 2], [3, 4], [5]]))
```

```python
# Q.2 - Nested Lists - Access a Nested Element
# Uses a list of indices to dig into a nested list
def access_nested_element(lst, indices):
    """
    Accesses a specific element in a nested list using indices.

    Parameters:
        lst (list): A nested list
        indices (list): List of indices to navigate to the element

    Returns:
        The element at the specified position
    """
    result = lst
    for i in indices:
        result = result[i]
    return result

print("Q.2:", access_nested_element([[1,2,3],[4,5,6],[7,8,9]], [1, 2]))
```

```python
# Q.3 - Nested Lists - Sum of All Elements
# Recursively adds all numbers regardless of how deep they are nested
def sum_nested(lst):
    """
    Calculates the sum of all elements in a nested list.

    Parameters:
        lst (list): A nested list of numbers

    Returns:
        float: Sum of all elements
    """
    total = 0
    for item in lst:
        if isinstance(item, list):  # if item is itself a list, go deeper
            total += sum_nested(item)
        else:
            total += item
    return total

print("Q.3:", sum_nested([[1, 2], [3, [4, 5]], 6]))
```

```python
# Q.4 - Nested Lists - Remove a Specific Element
# Goes through each sublist and removes all occurrences of elem
def remove_element(lst, elem):
    """
    Removes all occurrences of a specific element from a nested list.

    Parameters:
        lst (list): A nested list
        elem: The element to remove

    Returns:
        list: Nested list with the element removed
    """
    result = []
    for sublist in lst:
        result.append([item for item in sublist if item != elem])
    return result

print("Q.4:", remove_element([[1, 2], [3, 2], [4, 5]], 2))
```

```python
# Q.5 - Nested Lists - Find Maximum Element
# Recursively searches through all levels to find the biggest number
def find_max(lst):
    """
    Finds the maximum element in a nested list.

    Parameters:
        lst (list): A nested list of numbers

    Returns:
        float: The maximum value found
    """
    max_val = None
    for item in lst:
        if isinstance(item, list):  # if item is a list, go deeper
            candidate = find_max(item)
        else:
            candidate = item
        if max_val is None or candidate > max_val:
            max_val = candidate
    return max_val

print("Q.5:", find_max([[1, 2], [3, [4, 5]], 6]))
```

```python
# Q.6 - Nested Lists - Count Occurrences of an Element
# Counts how many times elem appears across all sublists
def count_occurrences(lst, elem):
    """
    Counts how many times an element appears in a nested list.

    Parameters:
        lst (list): A nested list
        elem: The element to count

    Returns:
        int: Number of times elem appears
    """
    count = 0
    for sublist in lst:
        for item in sublist:
            if item == elem:
                count += 1
    return count

print("Q.6:", count_occurrences([[1, 2], [2, 3], [2, 4]], 2))
```

```python
# Q.7 - Nested Lists - Deep Flatten (any depth)
# Like Q.1 but works no matter how deeply nested the list is
def deep_flatten(lst):
    """
    Flattens a deeply nested list into a single list regardless of depth.

    Parameters:
        lst (list): A deeply nested list

    Returns:
        list: A completely flattened single list
    """
    result = []
    for item in lst:
        if isinstance(item, list):  # if item is still a list, go deeper
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result

print("Q.7:", deep_flatten([[[1, 2], [3, 4]], [[5, 6], [7, 8]]]))
```

```python
# Q.8 - Nested Lists - Average of All Elements
# Reuses deep_flatten to get all numbers, then calculates average
def average_nested(lst):
    """
    Calculates the average of all elements in a nested list.

    Parameters:
        lst (list): A nested list of numbers

    Returns:
        float: The average of all elements
    """
    all_elements = deep_flatten(lst)
    return sum(all_elements) / len(all_elements)

print("Q.8:", average_nested([[1, 2], [3, 4], [5, 6]]))
```

```python
# SECTION: NumPy - Problem 4 - Matrix Operations
import numpy as np

# Given arrays (do not change these!)
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])


# -----------------------------------------------


# Q.1 - Matrix Operations - Prove A × A_inverse = Identity Matrix
# np.linalg.inv() gives the inverse of a matrix
# When you multiply a matrix by its inverse you always get identity matrix
A_inv = np.linalg.inv(A)
result1 = np.dot(A, A_inv)
print("Q.1 - A × A_inverse (should be Identity matrix):")
print(np.round(result1))          # np.round cleans up tiny decimal errors
```

```python
# Q.2 - Matrix Operations - Prove AB is NOT equal to BA
# Matrix multiplication is not commutative!
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\nQ.2 - AB:\n", AB)
print("      BA:\n", BA)
print("      AB == BA?", np.array_equal(AB, BA))    # should print False
```

```python
# Q.3 - Matrix Operations - Prove (AB)^T = B^T × A^T
# .T gives the transpose of a matrix (flips rows and columns)
AB_transpose = np.dot(A, B).T
BT_AT = np.dot(B.T, A.T)
print("\nQ.3 - (AB)^T:\n", AB_transpose)
print("      B^T × A^T:\n", BT_AT)
print("      Are they equal?", np.array_equal(AB_transpose, BT_AT))
```

```python
# Q.4 - Matrix Operations - Solve System of Linear Equations
# The equations are:
#  2x - 3y +  z = -1
#   x -  y + 2z = -3
#  3x +  y -  z =  9
#
# We write this as AX = B where:
# A = coefficient matrix, X = [x,y,z], B = result matrix

# Step 1: Write the equations as matrices
A_eq = np.array([[2, -3,  1],
                 [1, -1,  2],
                 [3,  1, -1]])

B_eq = np.array([-1, -3, 9])

# Step 2: Solve using inverse method  X = A_inverse × B
A_eq_inv = np.linalg.inv(A_eq)
X = np.dot(A_eq_inv, B_eq)
print("\nQ.4 - Solution using Inverse Method:")
print(f"      x = {round(X[0])}")
print(f"      y = {round(X[1])}")
print(f"      z = {round(X[2])}")
```

```python
# Q.5 - Matrix Operations - Solve Same Equations using np.linalg.solve
# np.linalg.solve is cleaner and more accurate than inverse method
X2 = np.linalg.solve(A_eq, B_eq)
print("\nQ.5 - Solution using np.linalg.solve:")
print(f"      x = {round(X2[0])}")
print(f"      y = {round(X2[1])}")
print(f"      z = {round(X2[2])}")
```
